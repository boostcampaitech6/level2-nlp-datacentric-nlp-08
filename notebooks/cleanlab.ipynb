```python
import torch
import pandas as pd
from tqdm import tqdm

from cleanlab import Datalab
from cleanlab.filter import find_label_issues
```

```python
output = pd.read_csv('../output.csv')
dataset_train = pd.read_csv('../data/train.csv')
```

```python
output = pd.read_csv('../output.csv')
probs_list = []

for idx, sample in tqdm(output.iterrows()):
    l = [float(x) for x in sample['prob'][1:-1].split(' ') if x != '']
    probs_list.append(l)  

probs_tensor = torch.Tensor(probs_list)
print(probs_tensor.shape)
```

```python
lab = Datalab(data=dataset_train, label_name="target")
lab.find_issues(pred_probs=probs_tensor.numpy())

lab.report()
lab.get_issues()
```

```python
from cleanlab.dataset import health_summary
class_names=[0,1,2,3,4,5,6]
health_summary(dataset_train['target'], probs_tensor.numpy(), class_names=class_names)
```

```python
ordered_label_issues = find_label_issues(
    labels=dataset_train['target'],
    pred_probs=probs_tensor.numpy(),
    return_indices_ranked_by='self_confidence',
)

head_issues=ordered_label_issues[:3]
for issue in head_issues:
    print('text', dataset_train.iloc[issue]['text'])
    print('label', dataset_train.iloc[issue]['target'])
```
